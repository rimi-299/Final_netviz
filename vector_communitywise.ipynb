```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your CSV file
df = pd.read_csv("cluster_info_linux_withoutcap.csv")
```

```python
#Reading the coordinates and force
data = pd.read_csv("coordinates_linux.csv",names=["Node","x","y"])
```

```python
# Filter out the outlier cluster
df = df[df['clusters'] != -1]
```

```python
#Top 20% edges

# Calculate the threshold value for the top 20% communities
threshold_value = df['Cluster_count'].quantile(0.9)

# Filter out communities with Cluster_count greater than the threshold
top_10_percent_communities = df[df['Cluster_count'] > threshold_value]

# Display the top 20% communities DataFrame
top_10_percent_communities
```

```python
edges_df = pd.read_csv("web-BerkStan.txt", delimiter='\t')

nodes_df = pd.read_csv("coordinates_with_clusters_linux.csv")
```

```python
edges_df
```

```python
top_10_percent_nodes = pd.merge(nodes_df, top_10_percent_communities[['clusters']], left_on='cluster', right_on='clusters')

top_10_percent_coordinates = top_10_percent_nodes[['node', 'n_x', 'n_y']]
```

```python
top_10_percent_nodes
```

```python
top_10_percent_coordinates
```

```python
merged_nodes_coordinates = pd.merge(top_10_percent_nodes, top_10_percent_coordinates, left_on='node', right_on='node')
```

```python
merged_nodes_coordinates
```

```python
# Merge cluster information with edges_df for source nodes
edges_cluster_merged = pd.merge(edges_df, merged_nodes_coordinates[['node', 'cluster']], left_on='Source', right_on='node', how='left')
edges_cluster_merged.rename(columns={'cluster': 'Source_cluster'}, inplace=True)

# Merge cluster information with edges_cluster_merged for target nodes
edges_cluster_merged = pd.merge(edges_cluster_merged, merged_nodes_coordinates[['node', 'cluster']], left_on='Target', right_on='node', how='left')
edges_cluster_merged.rename(columns={'cluster': 'Target_cluster'}, inplace=True)

# Filter edges to include only connections between nodes in different clusters
filtered_edges_different_cluster = edges_cluster_merged[
    edges_cluster_merged['Source_cluster'] != edges_cluster_merged['Target_cluster']
]

# Drop redundant columns
filtered_edges_different_cluster.drop(['node_x', 'node_y'], axis=1, inplace=True)
filtered_edges_different_cluster
```

```python
# Drop rows with missing cluster values
edges_cluster_merged.dropna(subset=['Source_cluster', 'Target_cluster'], inplace=True)

# Filter edges to include only connections between nodes in different clusters
filtered_edges_different_cluster = edges_cluster_merged[
    edges_cluster_merged['Source_cluster'] != edges_cluster_merged['Target_cluster']
]

# Drop redundant columns
filtered_edges_different_cluster.drop(['node_x', 'node_y'], axis=1, inplace=True)
filtered_edges_different_cluster
```

```python
filtered_edges_different_cluster
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the nodes
fig, ax = plt.subplots(figsize=(30, 25))
plt.scatter(merged_nodes_coordinates['n_x_x'], merged_nodes_coordinates['n_y_x'], color='skyblue', s=0.02)

# Plot edges between nodes
for _, edge in filtered_edges_different_cluster.iterrows():
    source_node = merged_nodes_coordinates[merged_nodes_coordinates['node'] == edge['Source']]
    target_node = merged_nodes_coordinates[merged_nodes_coordinates['node'] == edge['Target']]
    plt.plot([source_node['n_x_x'].values[0], target_node['n_x_y'].values[0]], 
             [source_node['n_y_x'].values[0], target_node['n_y_y'].values[0]], 
             color='red', linewidth=0.005)

plt.title('Edges Between Nodes in Different Clusters')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
